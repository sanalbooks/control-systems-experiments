# Nichols Chart: Frequency Response and Stability Analysis

## Introduction

The **Nichols chart** is a graphical representation of a system's **frequency response**, showing both the **magnitude** and **phase** of the system at different frequencies. It combines the information of the **Bode plot** (magnitude and phase) into a single plot, making it useful for analyzing the system’s stability and robustness.

The chart is used primarily in **control systems** to assess the system's **gain and phase margins** and to evaluate the overall system stability.

---

## Key Concepts

### 1. **Frequency Response**

The **frequency response** of a system describes how the system reacts to different frequency inputs. It is given by:

$$
G(j\omega) = \frac{Output}{Input}
$$

Where:
- $G(j\omega)$ is the **transfer function** of the system evaluated at the frequency $\omega$.
- $\omega$ is the **frequency** in rad/s.

The **magnitude** and **phase** of the system's frequency response are essential for understanding the system's behavior over a range of frequencies.

### 2. **Nichols Chart Representation**

The Nichols chart plots:
- **Phase** on the **x-axis** (ranging from $-180^\circ$ to $+180^\circ$).
- **Magnitude** (in **dB**) on the **y-axis** (ranging typically from $-40 \, \text{dB}$ to $+40 \, \text{dB}$).

The **phase** shows how much the output signal is delayed or advanced compared to the input, while the **magnitude** shows how much the output is scaled relative to the input at each frequency.

---

## **Nichols Chart Construction**

1. **Define the Transfer Function**:  
   For a system with a transfer function $G(s)$, evaluate the frequency response $G(j\omega)$ over a range of frequencies $\omega$ using:

   $$
   G(j\omega) = \frac{1}{m\omega^2 + c\omega + k}
   $$

   Where:
   - $m$ is the mass,
   - $c$ is the damping coefficient,
   - $k$ is the spring constant.

2. **Plot Phase vs Magnitude**:  
   For each frequency, plot the corresponding phase (in degrees) against the magnitude (in dB) to construct the Nichols chart.

---

## **Key Analysis Techniques**

### 1. **Stability Analysis Using Nichols Chart**

- **Stability Criterion**:  
  The stability of the system can be analyzed by inspecting how the Nichols plot interacts with the **critical point** $(0 \, \text{dB}, -180^\circ)$, which represents the point where the system would go unstable.
  
  - **Stable System**: If the Nichols plot does not encircle or cross the critical point $(0 \, \text{dB}, -180^\circ)$, the system is considered stable.
  - **Unstable System**: If the plot encircles or crosses the critical point, the system is at risk of becoming unstable.

### 2. **Gain Margin (GM)**

- The **gain margin** is defined as the distance from the **critical point** $(0 \, \text{dB}, -180^\circ)$ to the Nichols plot at **180° phase**.
- The **larger** the gain margin, the more **stable** the system is. If the plot touches or crosses the $0 \, \text{dB}$ line at $180^\circ$ phase, the system becomes unstable.

### 3. **Phase Margin (PM)**

- The **phase margin** is the phase at which the magnitude of the system is $0 \, \text{dB}$.
- The **larger** the phase margin, the more **robust** the system is to changes in gain or other system parameters.

---

## **Key Insights from the Nichols Chart**

| Parameter               | Meaning                                       |
|-------------------------|-----------------------------------------------|
| **Gain Margin (GM)**     | The distance from the plot to $0 \, \text{dB}$ at $180^\circ$ phase. Larger gain margin means greater stability. |
| **Phase Margin (PM)**    | The phase at $0 \, \text{dB}$ magnitude. Larger phase margin means the system is more robust. |
| **Wide Loops**           | Indicates **low damping** with potential for oscillations or instability. |
| **Narrow Loops**         | Indicates **high damping**, leading to more stable behavior. |
| **Crossing 180° Phase**  | Indicates the system is **unstable** and may lead to oscillations. |

---

## **Nichols Chart Interpretation**

1. **Stable System**:  
   A stable system has its **Nichols plot** without any encirclements or crossings through the **critical point** $(0 \, \text{dB}, -180^\circ)$. The system can tolerate significant changes in gain or phase before becoming unstable.

2. **Unstable System**:  
   If the **Nichols plot** crosses the $0 \, \text{dB}$ line at **180° phase**, the system will likely become unstable and oscillate. This indicates a high risk of instability.

# Nichols Chart Experiment for Mass-Spring-Damper System

## Objective:
The objective of this experiment is to generate and analyze the **Nichols Chart** for a **mass-spring-damper system** with varying damping coefficients. The Nichols chart is a frequency domain plot that shows the relationship between the system’s **magnitude** and **phase** at different frequencies. This helps in understanding the stability and dynamic behavior of the system.

## Theory:
### Mass-Spring-Damper System Transfer Function:
A **mass-spring-damper system** can be represented by the following transfer function:

$$
G(s) = \frac{1}{ms^2 + cs + k}
$$

Where:
- $m$ is the mass of the object (kg)
- $c$ is the damping coefficient (N·s/m)
- $k$ is the spring constant (N/m)
- $s$ is the Laplace transform variable

### Nichols Chart:
A **Nichols Chart** is a plot of **magnitude (in dB)** versus **phase (in degrees)** as a function of frequency. This chart is useful in analyzing system stability and determining how the system responds to different frequencies. It also helps in observing how the damping coefficient influences system performance.

### Stability Analysis:
- **Gain Margin (GM)** and **Phase Margin (PM)** are important stability indicators for a system.
- The system is considered **stable** if the **gain margin** is positive (indicating the system's ability to handle some increase in gain without becoming unstable).

## Experiment Setup:
1. Define the system's parameters:
   - Mass ($m$) = 1 kg
   - Spring constant ($k$) = 2 N/m
2. Vary the damping coefficient ($c$) across multiple values: 0.001, 0.1, 1.0, 5.0, and 10.0.
3. For each damping coefficient:
   - Generate the transfer function for the mass-spring-damper system.
   - Plot the **Nichols chart** showing the relationship between **phase** and **magnitude** at various frequencies.

## Procedure:
1. Define the transfer function for each damping coefficient using the formula:

   $$ 
   G(s) = \frac{1}{ms^2 + cs + k}
   $$

2. Compute the **Bode plot** of the transfer function over a frequency range using the `scipy.signal.bode` function.
3. Adjust the phase to ensure it lies within the range of $[-180^\circ, 180^\circ]$.
4. Plot the **Nichols chart** with phase on the x-axis and magnitude on the y-axis.
5. Analyze the stability of the system by examining the gain margin and phase margin. If the system crosses the critical point ($0 \, \text{dB}$ and $-180^\circ$), it may be unstable.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Define mass-spring-damper system parameters
m = 1.0  # mass (kg)
k = 2.0  # spring constant (N/m)

# Function to plot Nichols chart for different values of damping coefficient 'c'
def plot_nichols_for_damping(c_values):
    plt.figure(figsize=(10, 8))
    
    for c in c_values:
        # Define the transfer function: G(s) = 1 / (ms^2 + cs + k)
        numerator = [1]
        denominator = [m, c, k]
        system = signal.TransferFunction(numerator, denominator)

        # Frequency range for evaluation
        w = np.logspace(-2, 2, 1000)
        w, mag, phase = signal.bode(system, w)
        
        # Convert phase to continuous (unwrap) and limit to -360 to 0
        phase = np.unwrap(phase * np.pi/180) * 180/np.pi
        phase = np.mod(phase + 180, 360) - 180  # Wrap to [-180, 180]
        
        plt.plot(phase, mag, label=f'c = {c}')
    
    plt.title('Nichols Chart of Mass-Spring-Damper System')
    plt.xlabel('Phase (degrees)')
    plt.ylabel('Magnitude (dB)')
    plt.grid(True)
    plt.legend()
    
    # Add Nichols grid if desired (requires control library)
    try:
        from control import nichols_grid
        nichols_grid()
    except ImportError:
        print("For Nichols grid lines, install the 'control' library: pip install control")
    
    plt.show()

# List of different damping coefficients to test
c_values = [0.001, 0.1, 1.0, 5.0, 10.0]

# Plot the Nichols chart for different damping values
plot_nichols_for_damping(c_values)
```

## Observations:
1. **Effect of Damping on the Nichols Chart:**
   - When the damping coefficient ($c$) is very low (e.g., $c = 0.001$), the system shows a sharp peak in the magnitude, indicating that the system might be on the verge of instability. The phase is more negative, suggesting oscillatory behavior.
   - As the damping coefficient increases (e.g., $c = 0.1$, $c = 1.0$), the magnitude peak becomes less sharp, and the phase moves towards more stable values. This suggests that the system is becoming more stable with higher damping.
   - For high damping values (e.g., $c = 10.0$), the system exhibits a much lower peak in the magnitude plot and a more stable phase response. This indicates a well-damped system with reduced oscillations.

2. **Phase and Magnitude Relationship:**
   - The phase starts becoming more negative as the frequency increases, which is typical for underdamped systems.
   - As damping increases, the magnitude at certain frequencies reduces, indicating a decrease in the system's oscillatory response.
   - The system with lower damping tends to have a higher magnitude at certain frequencies, indicating higher amplification of certain input frequencies, which can lead to instability.

3. **Gain Margin and Phase Margin:**
   - Systems with lower damping tend to have a gain margin that approaches zero or becomes negative, suggesting that they are more susceptible to instability. 
   - Systems with higher damping have a positive gain margin, indicating that they are more stable and can tolerate higher gains without becoming unstable.

## Results:
- The **Nichols chart** allows us to visually assess the stability of the system for different damping coefficients. 
- **For low damping coefficients**, the system is more prone to instability, characterized by high peaks in the magnitude and a negative phase shift. These systems may oscillate more and have a smaller stability margin.
- **As damping increases**, the system becomes more stable with reduced oscillatory behavior. The magnitude peak decreases, and the phase response improves, moving towards more stable values.
- **Higher damping values** lead to a more stable system with lower peak magnitudes, implying that the system is less responsive to high-frequency inputs but is less prone to instability.
- The **Nichols chart** provides valuable insight into the system's behavior in the frequency domain, enabling us to adjust parameters like damping to achieve the desired performance.

In conclusion, by analyzing the Nichols chart for different damping coefficients, we can determine the stability of the system and the impact of damping on its frequency response. A system with adequate damping ensures stability and minimizes the risk of oscillations or instability.
