## What is a Ramp Response?

In control systems, the **ramp response** describes the system's output when the input is a **ramp function**. A ramp input is a signal that increases linearly over time, such as the increasing speed of a moving vehicle or an actuator whose displacement increases at a constant rate.

Mathematically, a ramp input is represented as:

$$
r(t) = \begin{cases}
0, & t < 0 \\
t, & t \geq 0
\end{cases}
$$

Where:
- \( r(t) \) is the ramp input, which increases linearly with time.

### Why Ramp Response?

The **ramp response** is used to analyze how a system behaves when the input increases at a constant rate, which is common in systems like motor speed control or variable position actuators. The ramp input is more challenging than a step input because it continuously increases, requiring the system to keep up with the changing input.

In control system analysis, the ramp response helps determine:
- **System's Ability to Track Changes**: How well the system can follow a continuously increasing input.
- **Steady-State Error**: In systems with a ramp input, steady-state error is used to assess how much the output deviates from the ramp input as time progresses.
  
The **ramp response** is often used to evaluate **type 1 systems**, which are systems that can respond to constant changes without accumulating error over time.

---

## Ramp Response vs Other Responses

- **Impulse Response**: Measures the system's response to an instantaneous input (idealized impulse).
- **Step Response**: Measures the system's response to a sudden change in input from zero to a constant.
- **Ramp Response**: Measures the system's response to a continuously increasing input, testing the system's ability to track changes over time.

The **ramp response** is important for evaluating how well a system can handle continuously changing inputs and tracking behavior over time.



## Experiment: Ramp Response of Mass-Spring-Damper System

### Objective:
In this experiment, we will analyze the **ramp response** of a **mass-spring-damper system**. The ramp response refers to the system's behavior when subjected to a ramp input, which increases linearly over time. This input simulates real-world scenarios such as increasing speed of a moving vehicle or actuator displacement at a constant rate.

### Introduction:
In control systems, the **ramp response** is crucial for understanding how well a system can track inputs that change continuously. Unlike step inputs, which are constant after a sudden change, ramp inputs grow steadily over time. The system's response to this type of input can provide insights into the system's ability to follow long-term, gradually increasing changes.

The mass-spring-damper system is a simple mechanical system often used to model damping effects. The governing differential equation for the system is:

$$
m \ddot{x}(t) + c \dot{x}(t) + k x(t) = F(t)
$$

Where:
- \( m \) is the mass,
- \( c \) is the damping coefficient,
- \( k \) is the spring constant, and
- \( x(t) \) is the displacement at time \( t \).

For this experiment, we use a **ramp function** as the input, which is mathematically represented as:

$$
r(t) = \begin{cases}
0, & t < 0 \\
t, & t \geq 0
\end{cases}
$$

The ramp input increases linearly with time, and we will observe the system's response as the ramp input is applied.

### Key Concepts:
- **Ramp Input**: A continuous, linear input function.
- **Ramp Response**: The system's output when a ramp input is applied. This is important for evaluating systems that need to track continuous changes.
- **Steady-State Error**: The error between the system's output and the ramp input at steady-state, which is particularly useful for evaluating the performance of type 1 systems.

### System Behavior:
The response of the system will vary based on the **damping coefficient** \( c \). The damping values used in this experiment will represent:
- **Underdamped System**: When \( c \) is small, the system might not follow the ramp input closely, and there could be oscillations.
- **Critically Damped System**: A system where \( c \) is chosen such that the system returns to steady-state without oscillating.
- **Overdamped System**: When \( c \) is large, the system will slowly follow the ramp input without oscillations.

### Procedure:
1. Define the system parameters: mass \( m \), spring constant \( k \), and damping coefficient \( c \).
2. Define the ramp input function: \( r(t) = t \).
3. Simulate the system's response using the **forced response** function in control systems.
4. Plot the **ramp response** for different damping values.

### Expected Outcome:
By analyzing the plots, we expect to see how the system's displacement increases over time in response to the ramp input, and how the system's ability to track the ramp changes for different damping values.

---

```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl

# System parameters
m = 1  # mass (kg)
k = 2  # spring constant (N/m)
damping_values = [0.5, 1.0, 3.0]  # Underdamped, Critically Damped, Overdamped
time = np.linspace(0, 10, 500)  # Time vector (from 0 to 10 seconds, 500 points)

# Ramp input function
def ramp_input(t):
    return t  # Ramp function: r(t) = t

# Plotting the ramp response for different damping values
plt.figure(figsize=(12, 6))

# Loop through each damping coefficient
for c in damping_values:
    # Create transfer function for each damping value
    system = ctrl.TransferFunction([1], [m, c, k])
    
    # Simulate ramp response by multiplying the ramp input with the system
    time_response, response = ctrl.forced_response(system, T=time, U=ramp_input(time))
    
    # Plot the response in time domain
    plt.subplot(1, 1, 1)
    plt.plot(time_response, response, label=f'Damping: c={c} Ns/m')

# Plotting the time-domain response
plt.title('Mass-Spring-Damper System Ramp Response for Different Damping')
plt.xlabel('Time (seconds)')
plt.ylabel('Displacement (meters)')
plt.legend()
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

## Analysis of Ramp Response

In the previous code, we simulated the **ramp response** of the mass-spring-damper system for three different damping values: underdamped, critically damped, and overdamped. Here's what we can observe:

1. **Time-Domain Response**: The plot shows how the system reacts to the ramp input. Unlike step inputs, the output continuously increases as the input ramps up. The system's ability to follow the ramp is reflected in how closely the output follows the input.

### Conclusion

The **ramp response** highlights the system's tracking capability and is especially important for evaluating systems with continuous changes in input. It is most useful for systems where steady-state error is a critical performance metric.
